```python
import quandl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open ("key.txt",'r') as f:
    key = f.read()
data = quandl.get(['WORLDBANK/ARM_TOT','WORLDBANK/LUX_TOT'], authtoken = key)

data.plot()
plt.xlabel("Years")
plt.ylabel("Terms of Trade")
plt.show()
```

```python
Arm_data=quandl.get("WORLDBANK/ARM_TOT", authtoken=key)
order_data = Arm_data.sort_values(by='Value', ascending=0)
max2=order_data['Value'][1]
print "2nd highest value for Armenia is  " + str(max2)
```

```python
Arm_exp = quandl.get("WTO/MERCH_EXP_ARM", authtoken=key)
Arm_imp = quandl.get("WTO/MERCH_IMP_ARM", authtoken=key)
Net_exp = Arm_exp ["Value"]-Arm_imp["Value"]
Net_exp.plot()
plt.ylabel("Net_Export")
```

```python
End_of_day_prices = quandl.get (["EOD/KO","EOD/MCD","EOD/DIS"], authtoken = key,trim_start= "1970-01-02", column_index = 4)
correlation = End_of_day_prices.corr()
print (correlation)
```

```python
import googlemaps
```

```python
with open ("api.txt",'r') as k:
    key2 = k.read()
gmaps=googlemaps.Client(key2)
```

```python
dist =[]
cities = ["Washington, DC","Los Angeles, CA", "Las Vegas, NV","Chicago, IL", "San Francisco, CA", "Dallas, TX","Garden City, KS", "Fort Collins, CO", "Miami, FL","Kansas City, MO"]

for i in cities:
        directions=gmaps.directions(i,"New York, NY", units = "metric")
        dist.append(directions[0]["legs"][0]["distance"])
        
for i in range(0,len(dist)):
    print('The distance between ' +  cities[i] + ' and ' + " New York, NY" + ' is ' + dist[i]["text"])
```
